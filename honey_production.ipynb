```python
df = pd.read_csv('honeyproduction.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
prod_per_year = df.groupby('year')['totalprod'].mean().reset_index()
```

### Modelling and Scatter Plot

```python
X = prod_per_year['year'].values.reshape(-1,1)
y = prod_per_year['totalprod']
```

```python
plt.scatter(X,y)
plt.show()
```

### Linear Regression Setup

```python
regr = LinearRegression()
```

```python
regr.fit(X,y)
"Coef: {}, Intercept: {}".format(regr.coef_,regr.intercept_)
```

```python
y_predict = regr.predict(X)
```

### Linear Regression

```python
plt.plot(X,y_predict)
plt.scatter(X,y)
```

### Linear Regression Prediction

```python
nums = np.array(range(2013,2051))
X_future = nums.reshape(-1,1)
```

```python
y_predict2 = regr.predict(X_future)
```

```python
plt.plot(X_future,y_predict2)
plt.scatter(X,y)
```
